```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns = iris.feature_names)
data['Species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Species'].value_counts()
```

```python
data.hist(bins=20, figsize=(7, 5), color="hotpink", edgecolor='yellow', grid=False)
plt.suptitle("Histogram", fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=data)
plt.title("Boxplot", fontweight='bold')
plt.show()
```

```python
sns.pairplot(data, hue='Species', markers=["o", "s", "D"], palette='viridis')
plt.suptitle("Pairplot", y=1.02, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.drop('Species', axis=1).corr(), annot=True, cmap='inferno', linewidths=.5)
plt.title("Heatmap of Feature Correlations", fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=data, palette='viridis')
plt.title("Countplot of Species", fontweight='bold')
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
